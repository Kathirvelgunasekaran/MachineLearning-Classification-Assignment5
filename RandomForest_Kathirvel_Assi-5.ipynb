```python
#Now we are going to convert all categorial data columns as Numerical numbers using One Hot Encoding 
dataset=pd.get_dummies(dataset,drop_first=True) # Nominal data to adrenal data get, while giving drop_first=True function it'll drop extra column.
dataset
```

```python
#Input and output split
#Data was already preprocessed. Now we are going to perform Input and output variables split
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']] 
indep #Now input column values seperately assigned into independent varaiable, Independent means input, Dependent means output.
```

```python
#Now 'charges' column value seperately assigned into dependent varaiable, dependent means output.
dep=dataset[["classification_yes"]] #Now charges value seperately assigned into dependent varaiable,
dep
#Now we got seperatley independent and dependent varaiables. now sucessfully splited input and output
```

```python
indep.shape
```

```python
#Split Trainset and Testset. 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep, dep, test_size=1/3, random_state=0) #No varaiable assigned here.here parallely split input and output based on 70:30 ratio.
#we are using _(underscore) instead of space into vraiables(X_train,X_test,y_train,y_test)
#X means input, y means output, X input based on trainset and testset(X_train,X_test) and also y was output based trainset and testset(y_train,y_test) 
#if you want to use sklearn library based on trainset and testset you have to split it like this four variables(X_train,X_test,y_train,y_test).
#As of now compltetd Trainset and Testset split.
#Trainset and Testset split is important for classification GridSearch(). many classification report, confusion_matrix,Type1Error and Type2Error all we have to seen. without trinset and testset split we can't prepare statistical report. score based only models will create.
```

```python
#Standardiztion Technique
#Now we are going try with standardization technique. we will see how it working with standardization technique. where we seperated Trainset and testset center we are going to try.
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler() #calling StandardScaler and assign to sc variable and function created
X_train=sc.fit_transform(X_train) #X_train varaiable sc(now algorithm in sc varaiable) "fit_transform(X_train)" -to find means value and standarization vale calculate for X_train and give that value.
X_test=sc.transform(X_test) #X_test varaiable sc(now algorithm in sc varaiable) "transform(X_test)" -"while using 'fit' means value and standarization vale calculate for X_train "which we calculated for (X_train) same like that it'll use same means value and standarization calculate value for X_test and gie that value.
#StandardScaler means defined range(-2 to +2)values, if we want to convert dataset values into defined range means(Example -2 to +2) without inormation loss. we can use this StandardScaler formula and convert for the same.
```

```python
#Model Creation
#ML algorithm all are provided by sklearn, we are creting this Model using trainset and tesset varaiable. 
from sklearn.model_selection import GridSearchCV #sklearn.model_selection import GridSearchCV, here GridSearchCV means create table and validate all the parameter values in multiple angle/combination. CV means cross validation.
#Using GridSerachCV(CV-cross validation) means create table in box and check all combination of parameters values and providing results in multiple angle like above.
#GridSerachCV--cross validation going to happen and it's related Gridsearch also going to perform here
from sklearn.ensemble import RandomForestClassifier # While calling, this line loaded into memory. here SVR may be function or class. In this line SVR considered as class
param_grid={'criterion':['gini','entropy'],
           'max_features': ['auto','sqrt','log2'],
            'n_estimators':[10,100]}
#playable paramerters using to check correct combination and provide result in tabular format.param_grid varaiable created and {} curley brace( {} dictionary) we can call it as dictionary. kernal means key('kernal':'rbf','poly','sigmoid','linear') and C means(10,100,1000,2000,3000).
#We have defined all the playable parameter values. Now going to call this model.
grid=GridSearchCV(RandomForestClassifier(),param_grid,refit=True, verbose=3,n_jobs=1,scoring='f1') #Here we calling GridSearchCV() and here we are passing SVR and passing parameters.refit=True means which one is best model take it as last model(). GridSearchCV() final output use it for best model. 'refit=False' means it will take last combination as (c:300, gmma:scale) like that only it'll take. 
#First four lines call it as procedure load. inside grid varaiable all models are there
#fitting the model for grid search
grid.fit(X_train,y_train) #fit function we are calling it using access operator, now passing splited trainset input and output.
#while executing above step fit() method model has been created. we got total model 120 model combination created.n_jobs=-1 means parallely running all models. once executed this it'll show all the executed parameter details.
#All models saved under grid varaiable.
```

```python
#print the best parameter after tunning
#print(grid.best_params_)
re=grid.cv_results_  #grid.cv_results() value stored in re varaiable. #here grid.cv_results() in this table stroed in dictionary format. readability will be poor,if it'sin table format means it will be good.
#print(re)
grid_predictions=grid.predict(X_test)  #grid_predictions(GridSearch()). Gridsearch  again used training dataset to display best parameter.
```

```python
#print classification report
#Classification means we have to use confusion_matrix
from sklearn.metrics import confusion_matrix #using confusion_matrix instead of r2_score function.
cm=confusion_matrix(y_test,grid_predictions) #confusion_matrix assigned to cm variable, now passing Actual output(y_test) and predicted output(grid_predictions)
#Test means always war between Actual data(y_test) and Evaluated data(grid_predictions)
```

```python
from sklearn.metrics import classification_report #using this confusion_matrix function going to find classification_report value.
clf_report=classification_report(y_test,grid_predictions) #under classification_report war against( Actual data(y_test) and Predicted data(y_pred))
#1.What is overall performance of RandomeForest- 0.98(98%)
#2.What kind of problem statement- Classification problem statement
#3.What is overall performance of your Model--Accuracy is overall percentage-- 0.98(98%)
#4.What is the Recall of 0-- 0.98(98%)
#5.What is the correct performance of CKD? --0.99(99%)
#6.Purchased-Correctly classified as well as wrongly classified-->(Model telling about whole value including wrongly classified)--0.99(99%)
#7.What is the f1-score of 0- 0.98(98%)
#8.What is the f1-score of 1- 0.99(99%)
#9.MacroAvg value for Precision--0.98(98%)
#10.MacroAvg value for Recall--0.98(98%)
#11.MacroAvg value for F1-Score--0.98(98%)
#12.Weighted average value for Precision--0.98(98%)
#13.Weighted average value for Recall--0.98(98%)
#14.Weighted average value for F1-Score--0.98(98%)
#15.Support value(Testdata count) -133
#16.It's Balanced dataset/Imbalanced dataset--balanced dataset
```

```python
#Just here showing for f1_score otherwise you can refer f1_score value in classification report also
from sklearn.metrics import f1_score #using this confusion_matrix function going to find classification_report value.
f1_macro=f1_score(y_test,grid_predictions,average='weighted') #under classification_report war against( Actual data(y_test) and Predicted data(y_pred))
print("The f1_macro value for best parameter{}:".format(grid.best_params_),f1_macro)
```
